```python
# Estatisticas descritivas das colunas ----->
# Exemplo: voltage (tensao em volts) e Global_intensity (vorrente em amperes)

# Media aritmetica
media_tensao = tabela["Voltage"].mean()
print(f"A media da tensao será de: {media_tensao:.2f}V")
# Mediana
mediana_tensao = tabela["Voltage"].median()
print(f"A mediana da tensao será de: {mediana_tensao:.2f}V")
# Desvio Padrão
desvio_padrao = tabela["Voltage"].std()
print(f"O desvio padrao da tensao será de: {desvio_padrao:.2f}V")
```

```python
# Media aritmetica
media_corrente = tabela["Global_intensity"].mean()
print(f"A media da corrente será de: {media_corrente:.2f}V")
# Mediana
mediana_corrente = tabela["Global_intensity"].median()
print(f"A mediana da corrente será de: {mediana_corrente:.2f}V")
# Desvio Padrão
desvio_corrente = tabela["Global_intensity"].std()
print(f"O desvio padrao da corrente será de: {desvio_corrente:.2f}V")
```

```python
# Adicionando uma coluna com Potencia aparente calculada
# Potencia Aparente = Tens"ao X Corrente
tabela["Pot_Calc_1"] = tabela["Voltage"] * tabela["Global_intensity"]
tabela.head()

#Pot_Calc_2 =  (Global_active_power ** 2 = Global_reactive_power ** 2) ** 0.5
```
